```python
import pandas as pd
import numpy as np
import scipy.io

def load_and_split_dataset(file_path, split_ratios, random_seed):

    if sum(split_ratios) != 1.0:
        raise ValueError("The sum of split_ratios must be 1.0")


    file_extension = file_path.split('.')[-1].lower()

    if file_extension == 'csv':

        data = pd.read_csv(file_path)
    elif file_extension == 'xls' or file_extension == 'xlsx':

        data = pd.read_excel(file_path)
    elif file_extension == 'mat':

        mat_data = scipy.io.loadmat(file_path)


        data = mat_data[next(iter(mat_data))]
    else:
        raise ValueError(f"Unsupported file extension: {file_extension}")

    if isinstance(data, pd.DataFrame):
        features = data.drop('Labels', axis=1)
        labels = data['Labels']
    else:

        features = data[:, :-1]
        labels = data[:, -1]


    if random_seed is not None:
        np.random.seed(random_seed)


    shuffled_indices = np.random.permutation(len(features))

    # Shuffle the data
    shuffled_features = features.iloc[shuffled_indices]
    shuffled_labels = labels.iloc[shuffled_indices]

    # Calculate split sizes
    total_size = len(features)
    train_size = int(split_ratios[0] * total_size)
    val_size = int(split_ratios[1] * total_size)
    test_size = total_size - train_size - val_size

    # Split the shuffled data
    X_train = shuffled_features[:train_size]
    y_train = shuffled_labels[:train_size]

    X_val = shuffled_features[train_size:train_size + val_size]
    y_val = shuffled_labels[train_size:train_size + val_size]

    X_test = shuffled_features[train_size + val_size:]
    y_test = shuffled_labels[train_size + val_size:]

    print(f"Training set shape: {X_train.shape} (Features), {y_train.shape} (Labels)")
    print(f"Validation set shape: {X_val.shape} (Features), {y_val.shape} (Labels)")
    print(f"Testing set shape: {X_test.shape} (Features), {y_test.shape} (Labels)")


    # Return the datasets
    return X_train, y_train, X_val, y_val, X_test, y_test

# Example usage:
file_path = "/content/Accidents_with_Label.csv"
X_train, y_train, X_val, y_val, X_test, y_test = load_and_split_dataset(file_path, (0.7, 0.15, 0.15), 42)
```
